```python
from pathlib import Path
import gtfs_kit as gk
import json
import datetime
DIR = Path('..')
sys.path.append(str(DIR))
DATA_DIR = DIR/'data/'
path = DATA_DIR/'CT_GTFS.zip'
feed = gk.read_feed(path, dist_units='km')
feed.describe()
feed.validate()
```

```python
# stations = {"University CTrain Station": "6816"}
stops_df = gk.stops.get_stops(feed)
# print(stops_df)
stops = []
for index, row in stops_df.iterrows():
    name = row["stop_name"]
    code = row["stop_code"]
    if "CTrain" in name:
        stop = (name, code)
        stops.append(stop)

print(stops)
```

```python
year = datetime.datetime.now().year
month = '{0:02d}'.format(datetime.datetime.now().month)
day = '{0:02d}'.format(datetime.datetime.now().day)
station_name = "University CTrain Station"
station = {station_name: "6816"}
dates = [str(year) + month + str(day)]
print(dates)
df = gk.build_stop_timetable(feed, station[station_name], dates)
times_list = []
for stop in stops: 
   for index, row in df.iterrows():
      departure = {}
      time = row["departure_time"].split(":")
      departure["year"] = year
      departure["month"] = month
      departure["day"] = day
      departure["hour"] = time[0]
      departure["minute"] = time[1]

      departure["line"] = "red"
      departure["station"] = stop[0]

      times_list.append(departure)

   filename = stop[0].replace(" ", "_").replace("/", "").lower()
   text_file = open("../data/sched/"+ filename + ".json", "w")
   text_file.write(json.dumps(times_list))
   text_file.close()
```
